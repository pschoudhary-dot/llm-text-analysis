```python
import pandas as pd
import spacy 
import textacy
import re
```

```python
train_data = pd.read_csv("train_essays.csv")
test_data = pd.read_csv("test_essays.csv")
train_prompt = pd.read_csv("train_prompts.csv")
sample_submissions = pd.read_csv("sample_submission.csv")
```

```python
train_data.head()
```

```python
def clean_text(text):
    text = re.sub(r'[^A-Za-z0-9\s.,!?\'"-]', '', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.lower()
    return text
```

```python
train_data['cleaned_text'] = train_data['text'].apply(clean_text)
clean_data=train_data.copy()
```

```python
clean_data.head()
```

```python
nlp = spacy.load('en_core_web_sm')
def process_text(text):
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return tokens
clean_data['tokens'] = clean_data['cleaned_text'].apply(process_text)
clean_data[['cleaned_text', 'tokens']].head()
```
